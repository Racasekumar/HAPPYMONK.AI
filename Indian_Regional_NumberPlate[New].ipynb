```python
NumberPlate_images_dict = {
    "Andaman_and_Nicobar" : list(data_dir.glob("Andaman_and_Nicobar/*")),
    "Andhra_Pradesh" : list(data_dir.glob("Andhra_Pradesh/*")),
    "Orissa" : list(data_dir.glob("Orissa/*")),
    "TamilNadu" : list(data_dir.glob("TamilNadu/*")),
       
}
```

```python
NumberPlate_images_dict["Orissa"]
```

```python
NumberPlate_labels_dict = {
    "Andaman_and_Nicobar" : 0,
    "Andhra_Pradesh" : 1,
    "Orissa" : 2,
    "TamilNadu" : 3,
       
}
```

```python
str(NumberPlate_images_dict["Orissa"][0])
```

```python
img = cv2.imread(str(NumberPlate_images_dict["Orissa"][0]))
img.shape
```

```python
cv2.resize(img,(180,180)).shape
```

```python
X, y = [], []

for number_plate , images in NumberPlate_images_dict.items():
    for image in images:
        img = cv2.imread(str(image))
        resized_img = cv2.resize(img,(180,180))
        X.append(resized_img)
        y.append(NumberPlate_labels_dict[number_plate])
```

```python
X = np.array(X)
y = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= 0)
```

```python
len(X_train), len(X_test)
```

```python
X_train_scaled = X_train / 255
X_test_scaled = X_test / 255
```

```python
num_classes = 4 

model = Sequential([
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(num_classes)  
])


model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
              
model.fit(X_train_scaled, y_train, epochs=50) 
```

```python
model.evaluate(X_test_scaled,y_test)
```
